# Do your own Exploratory Data Analysis

Exploratory Data Analysis (EDA) is a process in data analysis where one examines datasets to summarize their main characteristics, often using statistical graphics and other data visualization methods. EDA is crucial because it allows data analysts and scientists to gain insights into the patterns, anomalies, and relationships within the data before making any assumptions or building predictive models. This initial exploration can help in understanding the data's structure, detecting errors and outliers, identifying important variables, and formulating hypotheses for further analysis. In essence, EDA is fundamental in guiding the direction of further data analysis and ensuring that subsequent conclusions and models are based on a thorough understanding of the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Cuisine_rating.csv")
```

```python
df
```

```python
df.info()
```

```python
df.nunique()
```

```python
# Looping the columns and creating countplots for every column with 10 or less unique values

for i in df.columns:
    if df[i].nunique() <= 10:
        ax = sns.countplot(data=df, x=i)
        ax.set_xticklabels(labels=ax.get_xticklabels(), rotation=45, ha="center")
        plt.show()
```

```python
for i in df.columns:
    if df[i].nunique() <= 10:
        for j in range(1,6):
            ax = sns.countplot(data=df.loc[df["Food Rating"] == j], x=i)
            ax.set_xticklabels(labels=ax.get_xticklabels(), rotation=45, ha="center")
            plt.show()
```

```python
for i in df.columns:
    if df[i].nunique() <= 10:
        for j in range(1,6):
            ax = sns.countplot(data=df.loc[df["Service Rating"] == j], x=i)
            ax.set_xticklabels(labels=ax.get_xticklabels(), rotation=45, ha="center")
            plt.show()
```
